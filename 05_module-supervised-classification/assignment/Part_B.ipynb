## Step B3: Data Preparation

### Data Splitting

```python
# Drop redundant columns
cols_to_drop = ['species_names', 'species_label'] 

df_iris_cleaned = df_iris.drop(cols_to_drop, axis=1)
df_iris_cleaned.head()
```

## Step B6: Decision Boundary Visualisation

We are supposed to plot the decision boundary for the best model, but since all models had perfect scores, we will plot for all 3 of them.

### Reduce to 2 features

```python
# Plot how the best model separates classes in 2D feature space
# We reduce to 2 feature for visualisation

X_2d = df_iris_cleaned[['petal length (cm)', 'petal width (cm)']]
y_2d = df_iris_cleaned['species']

X_train_2d, X_test_2d, y_train, y_test = train_test_split(X_2d, y_2d, test_size=0.2, random_state=42)

# We scale the features again
scaler = StandardScaler()
X_train_2d_sc = scaler.fit_transform(X_train_2d)
X_test_2d_sc = scaler.transform(X_test_2d)
```

### Plot decision boundary

```python
# Create reusable function

def plot_decision_boundary(model, X_2d, y, ax, title):

    # Create grid decision boundary (Grid)
    x_min, x_max = X_2d[:, 0].min() - 1, X_2d[:, 0].max() + 1
    y_min, y_max = X_2d[:, 1].min() - 1, X_2d[:, 1].max() + 1

    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                         np.arange(y_min, y_max, 0.02))

    # Predict over grid
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Style settings
    sns.set_theme(
        style="whitegrid",
        palette="deep",
        context="talk"
    )

    ax.contourf(xx, yy, Z, alpha=0.3)

    sns.scatterplot(
        x=X_train_2d_sc[:,0],
        y=X_train_2d_sc[:,1],
        hue=y_train,
        palette='Set1'
    )
    ax.set_title(title)
    ax.set_xlabel("Petal Length (cm)", fontsize=10, fontweight="bold")
    ax.set_ylabel("Petal Width (cm)", fontsize=10, fontweight="bold")
    
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

models = [
    ("Logistic Regression", lr_2d),
    ("Decision Tree", dt_2d),
    ("Random Forest", rf_2d)
]

for ax, (name, model) in zip(axes, models):

    plot_decision_boundary(
        model, 
        X_train_2d_sc,
        y_train,
        ax,
        name 
    )
plt.suptitle("Iris Decision Boundary", fontsize=15, fontweight="bold", y=1.05)
plt.tight_layout()
plt.show()
```
