We can see that there are missing values under columns "zip" and "twp" which is a a total of 80492 rows out of thr 663522 rows (about 12% of our rows). We would fill these rows intelligently instead of dropping them

### Step 3: Check for Outliers

We are going to check for strange coordinates (lag, lng)


```python
# Inspect range
df[['lat', 'lng']].describe()
```

For this database (Pennsylvanina area), roughly expect:

- **lat**: ~39.72° to 42.27°
- **lng**: ~-74.68° to -80.52°

```python
# Find suspicious values
df[(df['lat'] < 39.72) | (df['lat'] > 42.27)]
df[(df['lng'] < -80.52) | (df['lng'] > -74.68)]
```

```python
# Alternative method is
outliers = df[
    ~df['lat'].between(39.72, 42.27)|
    ~df['lng'].between(-80.52, -74.68)
]

print(outliers)
```

```python
# Remove clearly wrong rows
df = df[(df['lat'].between(39.72, 42.27)) & (df['lng'].between(-80.52, -74.68))]
```

### Step 4: Features Engineering

```python
# extract hour, day of week, and month from the timeStamp colum
df['hour'] = df['timeStamp'].dt.hour
df['day_of_week'] = df['timeStamp'].dt.day_name()
df['month'] = df['timeStamp'].dt.month
```

```python
# Visualize hourly trends
df.groupby('hour')['e'].count().plot()
```

```python
# Visualize Calls by Day of Week
df.groupby('day_of_week')['e'].count().plot(kind='bar') 
```

```python
# Visualize Calls by Township
df['twp'].value_counts().head(10).plot(kind='barh')
```

```python
# Visualize Monthly Trends
df.groupby('month')['e'].count().plot()
plt.show()
```

```python
# Plot a Heatmap (Day vs Hour)
import seaborn as sns


pivot = df.pivot_table(index='day_of_week', columns='hour', values='e', aggfunc='count')

sns.heatmap(pivot)
```
