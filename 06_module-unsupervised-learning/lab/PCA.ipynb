# Principal Component Analysis
---
- Simple ML algorithm
- **Unsupervised**: enables identification of structure in unlabeled data 
- Like eigen decomposition and SVD, enables lossy compression: 
    - To minimize both loss of precision and data footprint, first **_principal component_** contains most variance (data structure), second PC contains next most, and so on.

- Involves many linear algebra concepts already covered. , e.g.:
    - Norms
    - Orthogonal and identity matrices
    - Trace opeartor 

```python
# Import libraries
from sklearn import datasets

iris = datasets.load_iris()
```

```python
iris.data.shape
```

The iris dataset is made up of 150 observations (rows) and 4 features (columns).

```python
iris.get('feature_names')
```

The four columns correspond to four features which are measurements of flower anatomy — sepal length, seal width, petal length and petal width (all measured in cm).

Where:
- petal: The large, showy, and bright coloured part of the flower.
- sepal: green structures, similar to leaves and below the petal

```python
# first 6 rows
iris.data[0:6,:]    
```

We used indexing to get the first 6 rows, because it still a numpy array and not a DataFrame.


Now we import the PCA module and compress the 4 features down to 2 PCs.

```python
from sklearn.decomposition import PCA

# Fit the PCA model
pca = PCA(n_components=2) 
X = pca.fit_transform(iris.data)
```

```python
# Check shape after compression
X.shape
```

We see that we now have 2 components instead of the 4 separate features for each flower and the number of rows in unchanged.

```python
X[0:6,:]
```

```python
# Plotting the 2 PCs
import matplotlib.pyplot as plt

_ = plt.scatter(X[:, 0], X[:, 1])
```
