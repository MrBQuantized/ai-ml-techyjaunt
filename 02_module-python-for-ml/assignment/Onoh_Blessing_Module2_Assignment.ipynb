**Task 4**: The timeStamp column is currently a string. Convert it to datetime, then extract the hour of each call as a NumPy array. Use np.bincount() to count how many calls occurred in each hour (0 to 23).

````python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
hour_arr = df['timeStamp'].dt.hour.to_numpy()

# Number of calls per hour(0 to 23) using np.bincount()
hourly_calls = np.bincount(hour_arr, minlength=24)
print(f'Number of calls per hour: {hourly_calls.tolist()}')
````

Converting `timeStamp` to datetime and extracting `hour` allows analysis of call volume by time of day. The `np.bincount()` result gives a direct count for each hour from 0 to 23, making it easy to identify rush hours versus quiet periods.

**Task 5**: Use boolean masking on your hour_arr to find: how many calls occurred between 08:00 and 18:00 (daytime), and how many outside that range (nighttime).

````python
# Number of calls between 08:00 and 18:00
calls_mask = (hourly_calls >= 8) & (hourly_calls <=18)
within_rng = np.sum(calls_mask)
outside_rng = np.sum(~calls_mask)

print(f'Number of calls between 08:00 and 18:00: {within_rng}')
print(f'Number of calls outside range (nighttime): {outside_rng}')
````

**Observation for Task 5:**

The daytime/nighttime split uses the extracted hour array to compare call periods. The current code counts how many hour bins are in each range, rather than the total number of calls for those bins, so the actual analysis should use the sum of `hourly_calls[8:19]` for daytime and the sum of the remaining bins for nighttime.

### **Observation**

---

**Advantage of converting Pandas columns to NumPy arrays:** 

NumPy arrays are stored as contiguous blocks of typed memory, which makes element-wise operations faster than equivalent Pandas operations, especially on large datasets like this one (**663,522 rows**). NumPy also provides lower-level functions (`np.bincount`, `np.mean`, boolean masking) that operate directly on raw arrays without the overhead of index alignment or dtype inference. This makes NumPy more efficient for **pure numerical computation**, while Pandas is better for labeled data manipulation.

**When to prefer `np.mean()` over `df['col'].mean()`:**
- When the column is already a NumPy array and you want to avoid converting back to a Series.  
- When chaining multiple NumPy operations (e.g. variance or standard deviation calculations) where staying in NumPy reduces overhead.  
- When working in functions or pipelines that expect NumPy arrays.  
- `df['col'].mean()` is more convenient when handling missing values, since Pandas automatically ignores `NaN`s. In NumPy, you would need `np.nanmean()` to achieve the same behavior.

**Latitude statistics summary:**

After cleaning, the expected range would be closer to **~0.5 degrees**.

### **C4 — Seaborn: Statistical Visualisation**

Use Seaborn to create the following statistical visualisations. Seaborn works directly with DataFrames — pass the DataFrame and column names as arguments.

**Task 1**: Create a countplot showing the number of calls per department. Use the palette parameter to apply a colour palette of your choice.


````python
plt.figure(figsize=(8, 5))

sns.countplot(
    data=df, 
    x='dept',
    hue='dept',
    palette='Set2'
)
plt.title('Number of Calls per Department')
plt.xlabel('Department')
plt.ylabel('Number of Calls')

plt.xticks(rotation=45)
plt.show()
````

**Task 2**: Create a heatmap showing the number of calls by day of week (0=Monday, 6=Sunday) and hour of day. First create a pivot table using the day and hour columns.

````python
# Extract Day and Hour
df['day_of_week'] = df['timeStamp'].dt.dayofweek
df['hour']        = df['timeStamp'].dt.hour


# GroupbyDay and Hour, then count calls
pivot = df.groupby(['day_of_week', 'hour'])['e'].count().unstack(fill_value=0)
pivot_map = {
    0: 'Monday',
    1: 'Tuesday',
    2: 'Wednesday',
    3: 'Thursday',
    4: 'Friday',
    5: 'Saturday',
    6: 'Sunday'
}

pivot.index = pivot.index.map(pivot_map)

# Plot Heat Map
plt.figure(figsize=(14, 6))

sns.heatmap(
    pivot,
    cmap='YlOrRd')
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')
plt.title('Call Volume by Day and Hour')

plt.show()
````

**Task 3**: Create a boxplot comparing the latitude values across the three departments. What does the spread of latitude tell you about the geographic range of each department's calls?

````python
plt.figure(figsize=(9, 6))

sns.boxplot(
    data=df, 
    x='dept', 
    y='lat', 
    hue='dept',
    palette='Blues',
    legend=False
)

plt.xlabel('Department')
plt.ylabel('Latitude')
plt.title('Latitude Distribution Across Departments')

plt.show()
````

**Task 4**: Create a Seaborn countplot showing the top 5 reasons for EMS calls only. Filter the DataFrame first, then use value_counts() to identify the top 5 before plotting.

````python
# Filter only EMS calls
ems_df   = df[df['dept'] == 'EMS']

# Get top 5 EMS call reasons

top5_ems = ems_df['reason'].value_counts().head(5).index

# Keep rows with only top 5reasons
ems_top  = ems_df[ems_df['reason'].isin(top5_ems)]

# Plot
plt.figure(figsize=(12, 6))

sns.countplot(
    data=ems_top,
    y='reason',
    hue='reason',
    palette='Set2',
    legend=False,
    order=top5_ems
)

plt.xlabel('Number of Calls')
plt.ylabel('Reason')
plt.title('Top 5 Reasons for EMS Calls')
plt.show()
````

---

*Heatmap (call volume by day of week and hour of day):*  

**Peak activity:**

The highest call volumes occur consistently on **weekday afternoons, especially between 3 PM and 6 PM (hours 15–18)**. Friday afternoons are particularly busy, likely due to increased traffic, end-of-week fatigue, and higher travel activity as people start the weekend.

**Daytime weekday pattern:**  
From Monday to Friday, call volume rises sharply around **7–8 AM**, remains high through the afternoon, and gradually declines after 6 PM. This pattern aligns with typical work and commute hours, where traffic accidents, workplace incidents, and medical emergencies are more common.

**Weekend pattern:**  
Saturday and Sunday show a **flatter and slightly delayed peak**. Activity rises later in the morning, reflecting slower starts to the day, and the afternoon peak is less sharp than on weekdays. Saturday evenings (8–11 PM) show slightly higher activity than weekday evenings, possibly linked to social events and alcohol-related incidents.

**Overnight trough:** 
The lowest activity occurs between **2 AM and 5 AM** across all days. This is expected due to reduced movement, closed businesses, and most people being asleep. The drop is most consistent on weekday nights, while Friday and Saturday nights remain slightly more active.

**Implication:** 
These patterns suggest emergency services could improve efficiency by **allocating more resources during weekday afternoons and weekend evenings**, while scaling down overnight staffing when demand is consistently low.
